```python
# But wait... these are the case codes, but we want the texts right? Let's look them up in the "/data/casescontent/rulings/" folder
# First import the os library (which translates our Python instructions into a language the Operating System can understand)
import os

# Get all the files listed in "/data/casescontent/rulings/"
files_with_full_text_of_case = os.listdir('../data/casescontent/rulings/')
```

```python
# First create a folder to store all our consumer protection cases (the full texts) 
try:
    # Create target Directory
    os.mkdir('../data/consumerprotection')
    print("Directory created!") 
except FileExistsError:
    print("Directory already exists!")
```

```python
# Python library for copying files
from shutil import copyfile

# Filter for only the consumer protection cases
for file in files_with_full_text_of_case:
    # Remove everything in the filename that is not the celex number
    strippedfilename = file.replace("full_text_", "")
    strippedfilename = strippedfilename.replace(".txt", "")
    
    print(strippedfilename)
    # Check if the filename is a valid consumer protection celex number
    if strippedfilename in unique_consumer_protection_celex_numbers:
        # Copy the file to our new folder
        copyfile('../data/casescontent/rulings/'+file, '../data/consumerprotection/'+file)
```
